```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Data
df = pd.read_csv('Data_Benzene_Reduced.csv', sep=";", header=None)
df.shape
```

```python
df.shape
```

```python
df.head() # returns the first n rows for the object based on position, 5 is the default number
```

```python
df.rename(columns={0: 'benzene', 1: 'eq_ratio'}, inplace=True) # the two first columns are "special"
```

```python
df.head()
```

```python
# Task 1
## Heatmaps
df_benzene = df.groupby('benzene', as_index=False).mean() # regroup data by benzene
```

```python
df_benzene
```

```python
grid_shape = (6,8) # new grid size wrt statement due to NaNs
```

```python
# plots T by benzene
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(30, 30), sharey=True)
fig.set_facecolor('white')
# find min and max values to get the same scale
min_val = df_benzene.iloc[:,2:].min().min()
max_val = df_benzene.iloc[:,2:].max().max()
for index, value in enumerate(df_benzene["benzene"]):
    T_data = df_benzene[df_benzene["benzene"] == value].iloc[:,2:].to_numpy().reshape(grid_shape)
    sns.heatmap(T_data, cmap="YlOrBr", ax=ax.flatten()[index], vmin=min_val, vmax=max_val, annot=True, fmt=".2f")
    ax.flatten()[index].set_title(f"Grid temperatures for {value} benzene percentage")
    plt.savefig("T_benzene.png", dpi=300)
```

```python
df_phi = df.groupby('eq_ratio', as_index=False).mean() # regroup data by equiv. ratio
```

```python
df_phi
```

```python
# plots T by eq. ratio
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(30, 30), sharey=True)
fig.set_facecolor('white')
# find min and max values to get the same scale
min_val = df_phi.iloc[:,2:].min().min()
max_val = df_phi.iloc[:,2:].max().max()
for index, value in enumerate(df_phi["eq_ratio"]):
    T_data = df_phi[df_phi["eq_ratio"] == value].iloc[:,2:].to_numpy().reshape(grid_shape)
    sns.heatmap(T_data, cmap="YlOrBr", ax=ax.flatten()[index], vmin=min_val, vmax=max_val, annot=True, fmt=".2f")
    ax.flatten()[index].set_title(f"Grid temperatures for equivalence ratio of {value}")
    plt.savefig("T_phi.png", dpi=300)
```
