## Benzene 

```python
# Heatmaps
df_benzene = df.groupby('benzene', as_index=False).mean() # regroup data by benzene
```

```python
df_benzene
```

```python
grid_shape = (6,8) # new grid size wrt statement due to NaNs
```

```python
# plots T by benzene
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(30, 30), sharey=True)
fig.set_facecolor('white')
# find min and max values to get the same scale
min_val = df_benzene.iloc[:,2:].min().min()
max_val = df_benzene.iloc[:,2:].max().max()
for index, value in enumerate(df_benzene["benzene"]):
    T_data = np.flip(df_benzene[df_benzene["benzene"] == value].iloc[:,2:].to_numpy().reshape(grid_shape),axis=0)
    sns.heatmap(T_data, cmap="YlOrBr", ax=ax.flatten()[index],vmin=min_val, vmax=max_val, annot=True, fmt=".2f")
    ax.flatten()[index].set_title(f"Grid temperatures for {value} benzene percentage")
    
    
#plt.savefig("T_benzene_heatmap.png", dpi=300) # to save the figure(s)
```

```python
## histograms
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 20), sharey=True)
fig.set_facecolor('white')

for index, benzene_p in enumerate(df["benzene"].unique()):
    # array of temperatures for each benzene percentage
    T_data = np.flip(df[df["benzene"] == benzene_p].iloc[:,2:].to_numpy().flatten(),axis=0)
    # plot the histogram
    sns.histplot(T_data, ax=ax.flatten()[index], kde=False, bins=50)
    ax.flatten()[index].set_title(f"Temperature distribution for {benzene_p}% benzene")
    plt.savefig("T_benzene_hist.png", dpi=300)
```

# Task 2 : Clustering and dim. reduction

## PCA and scaling 

### Scaling

```python
df_xy = pd.read_csv('test_reduced.csv',index_col=0)
df_xy.head()
```

```python
# vast scaling : gives strange results do not use it at least until solved
df_xy_vast = (df_xy-np.mean(df_xy,axis=0))/(np.std(df_xy,axis=0)**2/np.mean(df_xy,axis=0)) #(vast scaling)
df_xy_vast.head()
```

```python
# standard scaling : is ok
df_xy_std = (df_xy-np.mean(df_xy,axis=0))/np.std(df_xy,axis=0)
df_xy_std.head()
```

### PCA

```python
# PCA on dataset sclaed with standard scaling method
from sklearn.decomposition import PCA

ds_xy_std = df_xy_std.to_numpy()

pca_std = PCA(n_components=3)
pca_std.fit(ds_xy_std)
A_std = pca_std.components_.T
L_std = pca_std.explained_variance_ratio_
Z_std = ds_xy_std @ A_std
explained_var_ratio_std = pca_std.explained_variance_ratio_

# plot of the explained variance by PCs and the cumulative variance ratio :  3 comp. explain up to 80% of the total variance
plt.plot(explained_var_ratio_std, label = "explained variance ratio")
plt.plot(np.cumsum(explained_var_ratio_std),label = "cumulative variance ratio")
plt.legend(loc="upper left")
x_ticks = np.array([i for i in range(0,3)])
plt.xticks(x_ticks)
plt.xlabel('nb of components')
plt.ylabel('Variance ratio')
```

```python
# scatter plot of the PCs : not very useful 
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
im = ax.scatter(Z_std[:,0], Z_std[:,1], Z_std[:,2],cmap=plt.cm.get_cmap('tab10', 3), alpha=0.5)
ax.set_xlabel('Z1')
ax.set_ylabel('Z2')
ax.set_zlabel('Z3')
fig.colorbar(im)
fig.tight_layout()
plt.show()
```

# Clustering

```python
from sklearn.cluster import KMeans

# clustering test before dimensionality reduction
kmeans  = KMeans(n_clusters=5,init='random')
kmeans.fit(ds_xy_std)
labels = kmeans.labels_

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
im = ax.scatter(ds_xy_std[:,0], ds_xy_std[:,1],ds_xy_std[:,4],c=labels, cmap=plt.cm.get_cmap('tab10', 5), alpha=0.5)
ax.set_xlabel('Benzene')
ax.set_ylabel('Equivalence ratio')
ax.set_zlabel('Temperature')
fig.colorbar(im)
plt.show()
```

```python
# clustering on dataset after PCA : this is good, the one to take
kmeans  = KMeans(n_clusters=4,init='random')
kmeans.fit(Z_std)
labels = kmeans.labels_
print(labels) # to see the results, gives the cluster for each 
```

```python
# cluster on PCs which seems ok
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
im = ax.scatter(Z_std[:,0], Z_std[:,1],Z_std[:,2],c=labels, cmap=plt.cm.get_cmap('tab10', 4), alpha=0.9) # plot clusters
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
fig.colorbar(im)
#plt.savefig('Clustering_after_PCA.png',dpi = 300)
plt.show()
#pd.DataFrame(Z_std, columns =['PC1','PC2','PC3']).to_csv('PCs.csv')
```

```python
from sklearn.metrics import davies_bouldin_score
# to verify the number of clusters, 5 was chosen and verified a posteriori but the reverse yields so let's say we 
# did it the other way around
k_array = np.arange(2,6)
db_score_array = np.zeros((k_array.size,))
for i,k in enumerate(k_array):
    kmeans = KMeans(n_clusters = k, init = 'random')
    kmeans.fit(Z_std)
    labels = kmeans.labels_
    db_score_array[i] = davies_bouldin_score(Z_std, labels)
    
n_clusters = k_array[np.argmin(db_score_array)]
print(n_clusters)

plt.ylim(1.2, 1.4)

plt.plot(k_array,db_score_array)
plt.show()
```

## Neural Network 

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split


data_reduced = pd.read_csv('test_reduced.csv')
#data_reduced_std = (data_reduced-np.mean(data_reduced,axis=0))/np.std(data_reduced,axis=0)


data_reduced = data_reduced.iloc[:,1:]
for i in range (5) :
    data_reduced.iloc[:,i] = data_reduced.iloc[:,i]/(max(data_reduced.iloc[:,i])) # to have between 0 and 1

label = np.array(data_reduced["Temperature"])
columns = list(data_reduced.columns)
data_reduced_array = np.array(data_reduced.drop("Temperature", axis=1))

delta_temp = max(label)-min(label)

train_data, test_data, train_labels, test_labels = train_test_split(data_reduced_array, label, test_size = 0.20, random_state=10)
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(4, activation='relu',input_dim=4))
model.add(tf.keras.layers.Dense(600, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='tanh'))
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt,
                loss='mean_squared_error',
                metrics=['MeanSquaredError'])
```

```python
history = model.fit(train_data, train_labels, 
                    epochs=8, 
                    validation_data=(test_data, test_labels))
```

```python
value = model.predict(test_data[0:10])
for i in range(len(value)):
    print(f"Predict : {abs((value[i]-test_labels[i])/(test_labels[i])*100)}% of error")
```
