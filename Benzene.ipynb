# Task 1 : data visualization and representation

## Benzene 

```python
# Heatmaps
df_benzene = df.groupby('benzene', as_index=False).mean() # regroup data by benzene
```

```python
df_benzene
```

```python
grid_shape = (6,8) # new grid size wrt statement due to NaNs
```

```python
# plots T by benzene
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(30, 30), sharey=True)
fig.set_facecolor('white')
# find min and max values to get the same scale
min_val = df_benzene.iloc[:,2:].min().min()
max_val = df_benzene.iloc[:,2:].max().max()
for index, value in enumerate(df_benzene["benzene"]):
    T_data = df_benzene[df_benzene["benzene"] == value].iloc[:,2:].to_numpy().reshape(grid_shape)
    sns.heatmap(T_data, cmap="YlOrBr", ax=ax.flatten()[index], vmin=min_val, vmax=max_val, annot=True, fmt=".2f")
    ax.flatten()[index].set_title(f"Grid temperatures for {value} benzene percentage")
    #plt.savefig("T_benzene_heatmap.png", dpi=300) # to save the figure(s)
```

```python
## histograms
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 20), sharey=True)
fig.set_facecolor('white')

for index, benzene_p in enumerate(df["benzene"].unique()):
    # array of temperatures for each benzene percentage
    T_data = df[df["benzene"] == benzene_p].iloc[:,2:].to_numpy().flatten()
    # plot the histogram
    sns.histplot(T_data, ax=ax.flatten()[index], kde=False, bins=50)
    ax.flatten()[index].set_title(f"Temperature distribution for {benzene_p}% benzene")
    #plt.savefig("T_benzene_hist.png", dpi=300)
```

```python
fig = plt.figure(figsize=(12,8))
# put all temperatures in a single column
benzene_boxplot_df = df.melt(id_vars=("benzene", "eq_ratio"), var_name="grid_position", value_name="temperature")
sns.boxplot(x="benzene", y="temperature", data=benzene_boxplot_df)
plt.xlabel('Benzene %')
plt.ylabel('Temperature [K]')
#plt.savefig("T_benz_boxplot.png", dpi=300)
```

# Task 2 : Clustering and dim. reduction

## PCA

```python
df_std = (df-np.mean(df,axis=0))/np.std(df,axis=0)
df_std.head()

```

```python
df_vast = (df-np.mean(df,axis=0))/(np.std(df,axis=0)**2/np.mean(df,axis=0))
df_vast.head()
```

```python
from sklearn.decomposition import PCA

ds_std = df_std.to_numpy()

pca_std = PCA(n_components=10)
pca_std.fit(ds_std)
A_std = pca_std.components_.T
L_std = pca_std.explained_variance_ratio_
Z_std = ds_std @ A_std
explained_var_ratio = pca_std.explained_variance_ratio_

plt.plot(explained_var_ratio)
```

```python
ds_vast = df_vast.to_numpy()

pca_vast = PCA(n_components=10)
pca_vast.fit(ds_vast)
A_vast = pca_vast.components_.T
L_vast = pca_vast.explained_variance_ratio_
Z_vast = ds_vast @ A_vast
explained_var_ratio_vast = pca_vast.explained_variance_ratio_

plt.plot(explained_var_ratio_vast)
```

```python
#to verify
fig = plt.figure(figsize=(10,5))
scat = plt.scatter(Z_vast[:,0], Z_vast[:,1], cmap=plt.cm.get_cmap('tab10', 3), alpha=0.5)
plt.colorbar(scat, ticks=[0,1,2])
```

```python
x_ticks = np.array([i for i in range(0,10)])

ds = df.to_numpy()
pca = PCA(n_components=10)
pca.fit(ds)
A = pca.components_.T
L = pca.explained_variance_ratio_
Z = ds @ A
fig = plt.figure(figsize=(10,3))
plt.plot(L, 'ko--', label='No scaling')
plt.plot(L_std, 'ro--', label='Standard (auto) scaling')
plt.plot(L_vast, 'bo--', label='VAST scaling')
plt.xticks(x_ticks)
plt.xlabel('Number of components')
plt.ylabel('Eigenvalue')
plt.legend()
```
